```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:/Users/dyane/Documents/personal_transactions.csv")
df.head()
```

```python
plt.boxplot(df['Amount'])
plt.yscale('log')
plt.show()
```

```python
plt.hist(df['Amount'],color='g')
plt.yscale('log')
plt.show()
```

```python
plt.figure(num=None, figsize=(40, 8), dpi=500, facecolor='w', edgecolor='k')
sns.boxplot(x='Category', y='Amount', data=df)
plt.yscale('log')
plt.show()
```

```python
sns.barplot(x='Account Name', y='Amount', data=df)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.barplot(x='Account Name', y='Amount', data=df)
plt.title("Average Sales")
plt.subplot(1, 2, 2)
sns.barplot(x='Account Name', y='Amount', data=df, estimator=np.median)
plt.title("Median Sales")
plt.show()
```

```python
plt.figure(figsize=(20, 8))
sns.barplot(x='Amount', y="Category", data=df, estimator=np.median)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.countplot(y="Category", data=df)
plt.show()
```
